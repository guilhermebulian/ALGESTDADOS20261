```python
#ATIVIDADE 04 - Crie um programa que peça um número inteiro e mostre seu antecessor e seu sucessor.
digito= int(input('Digite seu número; '))
print(f"sucessor: {digito+1}")
print(f"antecessor: {digito-1}")
```
