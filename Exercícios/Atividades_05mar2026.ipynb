```python
#ATIVIDADE 03 - Desenvolva um programa que peça dois números ao usuário e mostre a soma entre eles.
digito1 = int(input('Digite o primeiro número: '))
digito2 = int(input('Digite o segundo número: '))
print(f"{digito1} + {digito2} = {digito1 + digito2}")
```

```python
#ATIVIDADE 04 - Crie um programa que peça um número inteiro e mostre seu antecessor e seu sucessor.
digito= int(input('Digite seu número: '))
print(f"sucessor: {digito+1}")
print(f"antecessor: {digito-1}")
```

```python
#ATIVIDADE 05 - Faça um programa que peça o ano de nascimento do usuário e calcule sua idade aproximada.
ano= int(input('Digite seu ano de nascimento: '))
print(f"Sua idade aproximada é: {2026-ano}")
```

```python
#ATIVIDADE 06 - Desenvolva um programa que peça duas notas e calcule a média entre elas.
nota1= float(input('Digite sua primeira nota: '))
nota2= float(input('Digite sua segunda nota: '))
print(f"A média das suas duas notas ({nota1} e {nota2}) é igual a {(nota1+nota2)/2}")
```

```python
#ATIVIDADE 07 - Crie um programa que peça uma temperatura em graus Celsius e converta para Fahrenheit.
temp= float(input('Digite a temperatura da sua cidade (ºC): '))
print(f"A sua temperatura em Fahrenheit é: {temp*1.8+32}°F")
```

```python
#ATIVIDADE 08 - Faça um programa que peça a largura e a altura de um retângulo e calcule sua área.
larg= float(input('Digite a largura do seu retângulo: '))
alt= float(input('Digite a altura do seu retângulo: '))
print(f"A área total da superfície do seu retângulo é: {(larg*alt)}")
```

```python
#ATIVIDADE 09 - Desenvolva um programa que peça um número inteiro e calcule o seu quadrado.
num= int(input('Digite um número inteiro; '))
print(f"O quadrado do número {num} é igual a: {(num*num)}")
```

```python
#ATIVIDADE 10 - Crie uma calculadora simples que peça dois números e uma operação matemática (+, -, *, /) e mostre o resultado
num1 = input("Insira seu primeiro número: ")
operador = input('Insira operador (+, -, * ou /): ')
num2 = input("Insira seu segundo número: ")
resultado = eval(num1+operador+num2)
print(resultado)
```
