```python
# -- Digite e execute linha a linha -----
a = 10
b = 3
print(a + b) #Adição
print(a - b) #Subtração
print(a * b) # Multiplicação
print(a / b) #Divisão real
print(a // b) #Divisão Int.
print(a % b) #Módulo
print(a ** b) #Potenciação

#Qual a diferença entre / e // ?
#Por que a / b não dá 3 exato?
```

```python
# -- Execute linha a linha e preveja o resultado antes ---

nota = 7.5

print(nota >= 6.0) # ? (aprovado?)
print(nota == 10) # ? (nota máxima?)
print(nota != 5.0) # ? (diferente de 5?)
print(nota < 5.0) # ? (reprovado?)

# O que achou dos resultados? Refletiram a realidade?
# E se alternassemos a nota para 5.0?
```

```python
# -- Troque os valores e observe o que muda -----
idade = 17
renda = 1500.0

# and: as DUAS precisam ser True
print(idade >= 18 and renda >= 1200) # ?

# or : BASTA UMA ser True
print(idade >= 18 or renda >= 1200) # ?

# not: inverte
tem_debito = False
print(not tem_debito) # ?
```

```python
# -- Preveja o resultado de cada expressão -----
print(2 + 3 * 4) # ? → qual operador executa primeiro?
print((2 + 3) * 4) # ? → parênteses mudam tudo!
print(2 ** 3 ** 2) # ? → ** associa da DIREITA p/ esquerda
print(10 / 2 + 3) # ?
print(10 / (2 + 3)) # ? → mesma conta, resultado diferente!

#Expressão mista (aritmético + relacional + lógico):
x = 3 + 2 * 4 > 10 and not False
#: aritmético primeiro, depois relacional, depois lógico
print(x) # ?
```
